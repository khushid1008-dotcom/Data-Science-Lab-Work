```
# Problem 2: R² (Coefficient of Determination)

# Given values
SSR = 420
SSE = 180

# Total Sum of Squares
SST = SSR + SSE

# R squared calculation
R_squared = SSR / SST

# Print results
print("Total Sum of Squares (SST):", SST)
print("R² value:", R_squared)

# Interpretation
if R_squared > 0.7:
    print("The model explains more than 70% variance.")
else:
    print("The model explains less than 70% variance.")
```

```
# Problem 1: Confusion Matrix Metrics

# Given values
TP = 50
FP = 8
TN = 32
FN = 10

# Total predictions
total_predictions = TP + FP + TN + FN

# Metrics calculations
accuracy = (TP + TN) / total_predictions
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1_score = 2 * precision * recall / (precision + recall)

# Print results
print("Total Predictions:", total_predictions)
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1_score)
```

```
# Problem 3: Mean Squared Error (MSE)

# Squared errors
e1 = 4
e2 = 9
e3 = 16
e4 = 25
e5 = 36

# Mean Squared Error calculation
MSE = (e1 + e2 + e3 + e4 + e5) / 5

# Print result
print("Mean Squared Error (MSE):", MSE)
```

```
# Problem 4: Cook's Distance

# Given values
ei_square = 9
MSE = 3
p = 3
hii = 0.2
n = 25

# Cook's Distance formula
cooks_distance = (ei_square / (p * MSE)) * (hii / ((1 - hii) ** 2))

# Threshold
threshold = 4 / n

# Print results
print("Cook's Distance:", cooks_distance)
print("Threshold:", threshold)

if cooks_distance > threshold:
    print("The observation is influential.")
else:
    print("The observation is not influential.")
```

```
# Problem 5: Simple Linear Regression Prediction

# Given regression equation: y = mx + c
m = 2      # slope
c = 5      # intercept
x = 10     # rainfall value

# Prediction
y = m * x + c

# Print result
print("Predicted crop yield:", y)

# Explanation
print("Slope (m) represents change in yield per unit rainfall.")
print("Intercept (c) represents yield when rainfall is zero.")
```
